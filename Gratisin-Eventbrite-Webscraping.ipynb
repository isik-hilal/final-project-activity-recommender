### Getting the events under the categories

### Music

```python
url_music = 'https://www.gratis-in-berlin.de/musik'

gratisin_music = requests.get(url_music)
```

```python
gratisin_music = BeautifulSoup(gratisin_music.content, "html.parser")
```

```python
music_names=[]
for i in gratisin_music.select('h2', class_='singletip')[1:]:
    x=i.get_text(strip=True).lower()
    music_names.append(x)

music_names
```

```python
pd.set_option('display.max_rows', None)

music_names_dict = {'name':music_names}
music_names_df =pd.DataFrame (music_names_dict)
music_names_df = music_names_df[:69]
music_names_df
```

#### Getting the dates

```python
music_dates=[]
for i in gratisin_music.select('.value.field_Date'):
    b=i.get_text(strip=True).lower()[-10:-5]
    music_dates.append(b)

music_dates
```

```python
music_dates_dict = {'date':music_dates}
music_dates_df =pd.DataFrame(music_dates_dict)
music_dates_df = music_dates_df[:69]
music_dates_df
```

#### Getting the links of the music events

```python
music_links=[]
for i in gratisin_music.select('.singletip'):
    
    music_links.append(i['href'])
music_links
```

```python
music_links_dict = {'link':music_links}
music_links_df =pd.DataFrame(music_links_dict)
music_links_df =music_links_df[:69]
music_links_df
```

```python
music_gratisin = pd.concat([music_names_df, music_dates_df, music_links_df], axis=1 ) 

music_gratisin.index += 1
music_gratisin
```

```python
url_may = 'https://www.eventbrite.com/d/germany--berlin/free--music--events--this-month/?page=1'
response = requests.get(url_may)
may_music = BeautifulSoup(response.content, 'html.parser')
```

#### Getting the names of music events in may

```python
eb_music_names_may=[]
for i in may_music.select('div.eds-is-hidden-accessible')[1::2]:
    x=i.get_text(strip=True).lower()
    eb_music_names_may.append(x)

eb_music_names_may_dict = {'name':eb_music_names_may}
eb_music_names_may_df =pd.DataFrame(eb_music_names_may_dict)


eb_music_names_may_df
```

#### Getting the dates of the music events in may

```python
eb_music_dates_may=[]
for i in may_music.select('.eds-event-card-content__sub-title')[1::2]:
    x=i.get_text(strip=True)[9:11].lower()
    eb_music_dates_may.append(x)
eb_music_dates_may_dict = {'date':eb_music_dates_may}
eb_music_dates_may_df =pd.DataFrame(eb_music_dates_may_dict)
eb_music_dates_may_df['date'] = eb_music_dates_may_df['date'].str.replace(',','')
eb_music_dates_may_df['date'] = eb_music_dates_may_df['date'].astype(str)+'.05' 
eb_music_dates_may_df['date'] = eb_music_dates_may_df['date'].str.replace('8:.05','Tomorrow')
eb_music_dates_may_df['date'] = eb_music_dates_may_df['date'].str.replace('at.05','Tomorrow')

eb_music_dates_may_df
```

#### Getting the links of the music events in May

```python
eb_music_links_may=[]
for i in may_music.select('.eds-event-card-content__action-link')[1::4]:
    eb_music_links_may.append(i['href'])
eb_music_links_may

eb_music_links_may_dict = {'link':eb_music_links_may}
eb_music_links_may_df =pd.DataFrame(eb_music_links_may_dict)

eb_music_links_may_df
```

#### Preparing the dataframe


```python
eb_music_may = pd.concat([eb_music_names_may_df, eb_music_dates_may_df, eb_music_links_may_df], axis=1)
eb_music_may.index +=1

eb_music_may
```

```python
eb_sport_names_au = []
eb_sport_dates_au = []
eb_sport_links_au= []

for i in range(1,3):
    url = 'https://www.eventbrite.com/d/germany--berlin/free--sports-and-fitness--events/?end_date=2022-08-31&page='
    url = url + str(i) +'&start_date=2022-08-01'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    names = soup.select('div.eds-is-hidden-accessible')[1::2]
    dates = soup.select('.eds-event-card-content__sub-title')[1::2]
    links = soup.select('.eds-event-card-content__action-link')[1::4]
    for name in names:
        eb_sport_names_au.append(name.get_text(strip=True).lower())
    for date in dates:
        eb_sport_dates_au.append(date.get_text(strip=True)[9:11].lower())
    for link in links:
        eb_sport_links_au.append(link['href'])
```

```python
eb_sport_au_dict = {'name':eb_sport_names_au, 'date':eb_sport_dates_au, 'link':eb_sport_links_au}
# Convert dictionary to Pandas dataframe
eb_sport_au_df = pd.DataFrame(eb_sport_au_dict)
eb_sport_au_df.index = eb_sport_au_df.index + 1
eb_sport_au_df['date'] = eb_sport_au_df['date'].str.replace(',','')
eb_sport_au_df['date'] = eb_sport_au_df['date'].astype(str)+'.08'
eb_sport_au_df['date'] = eb_sport_au_df['date'].str.replace('8:.08','Tomorrow')
eb_sport_au_df['date'] = eb_sport_au_df['date'].str.replace('4:.08','Tomorrow')
eb_sport_au_df['date'] = eb_sport_au_df['date'].str.replace('5:.08','Tomorrow')
eb_sport_au_df['date'] = eb_sport_au_df['date'].str.replace('at.08','Tomorrow')
eb_sport_au_df
```
